```python
import pandas as pd
from typing import List

def join_datasets(file_paths: List[str]) -> pd.DataFrame:
    """
    Perform multiple join operations on Olist datasets.
    
    Parameters:
        file_paths (List[str]): List of CSV file paths containing specific Olist datasets.
        
    Returns:
        pd.DataFrame: A DataFrame after performing all the specified joins.
    """
    
    # Step 3: Load CSV files into DataFrames
    dataframes = {}
    for file_path in file_paths:
        key = file_path.split("/")[-1].replace(".csv", "")
        dataframes[key] = pd.read_csv(file_path)
    
    # Step 4: Join datasets
    # Join olist_order_items_dataset to olist_orders_dataset
    merged_df = pd.merge(dataframes['olist_orders_dataset'], dataframes['olist_order_items_dataset'], on='order_id', how='left')
    
    # Join olist_order_payments_dataset to olist_orders_dataset
    merged_df = pd.merge(merged_df, dataframes['olist_order_payments_dataset'], on='order_id', how='left')
    
    # Join olist_order_reviews_dataset to olist_orders_dataset
    merged_df = pd.merge(merged_df, dataframes['olist_order_reviews_dataset'], on='order_id', how='left')
    
    # Join olist_customers_dataset to olist_orders_dataset
    merged_df = pd.merge(merged_df, dataframes['olist_customers_dataset'], on='customer_id', how='left')
    
     #'seller_zip_code_prefix' 'customer_zip_code_prefix'
    
    # Join olist_sellers_dataset to olist_orders_dataset
    merged_df = pd.merge(merged_df, dataframes['olist_sellers_dataset'], on='seller_id', how='left')

        # Join olist_geolocation_dataset to olist_customers_dataset
    merged_df = pd.merge(merged_df, dataframes['olist_geolocation_dataset'], 
                         left_on='seller_zip_code_prefix', 
                         right_on='geolocation_zip_code_prefix', how='left')
    
    return merged_df

result_df = join_datasets(csv_files)

result_df
```
